# 05 - LogReg Smoothed Results

Visualize results from `05_logreg_smoothed.py`.
Compares smoothed variants against the basis-only (no smoothing) baseline.

```python
import pandas as pd
import plotly.graph_objects as go

from _common import RESULTS_DIR

PLOT_LINESTYLES = {
    "none": "solid",
    "savgol": "dash",
    "gaussian": "dashdot",
    "spline": "dot",
}

raw = pd.read_csv(RESULTS_DIR / "logreg_smoothed_raw.csv")
summary = pd.read_csv(RESULTS_DIR / "logreg_smoothed_summary.csv")

raw = raw[raw["basis"] != "fourier"]
summary = summary[summary["basis"] != "fourier"]

print(f"Raw rows: {len(raw)}, Summary rows: {len(summary)}")
summary.head()
```

## PR-AUC vs K -- one panel per basis, lines by smoothing method

```python
bases = sorted(summary["basis"].unique())

for basis in bases:
    sub = summary[summary["basis"] == basis].sort_values("n_coeffs_per_arm")
    fig = go.Figure()
    for smoothing, grp in sub.groupby("smoothing"):
        fig.add_trace(go.Scatter(
            x=grp["n_coeffs_per_arm"], y=grp["pr_auc_mean"],
            error_y=dict(type="data", array=grp["pr_auc_std"].values, visible=True),
            mode="lines+markers",
            name=smoothing,
            line=dict(dash=PLOT_LINESTYLES.get(smoothing, "solid")),
        ))
    fig.update_layout(
        title=f"PR-AUC vs K — {basis}",
        xaxis_title="K (coefficients per arm)",
        yaxis_title="PR-AUC (mean)",
        height=400,
        template="plotly_white",
    )
    fig.show()
```

## ROC-AUC vs K -- one panel per basis

```python
for basis in bases:
    sub = summary[summary["basis"] == basis].sort_values("n_coeffs_per_arm")
    fig = go.Figure()
    for smoothing, grp in sub.groupby("smoothing"):
        fig.add_trace(go.Scatter(
            x=grp["n_coeffs_per_arm"], y=grp["roc_auc_mean"],
            error_y=dict(type="data", array=grp["roc_auc_std"].values, visible=True),
            mode="lines+markers",
            name=smoothing,
            line=dict(dash=PLOT_LINESTYLES.get(smoothing, "solid")),
        ))
    fig.update_layout(
        title=f"ROC-AUC vs K — {basis}",
        xaxis_title="K (coefficients per arm)",
        yaxis_title="ROC-AUC (mean)",
        height=400,
        template="plotly_white",
    )
    fig.show()
```

## Precision vs K -- one panel per basis, lines by smoothing method

```python
for basis in bases:
    sub = summary[summary["basis"] == basis].sort_values("n_coeffs_per_arm")
    fig = go.Figure()
    for smoothing, grp in sub.groupby("smoothing"):
        std_col = "precision_std" if "precision_std" in grp.columns else None
        fig.add_trace(go.Scatter(
            x=grp["n_coeffs_per_arm"], y=grp["precision_mean"],
            error_y=dict(type="data", array=grp[std_col].values, visible=True) if std_col else None,
            mode="lines+markers",
            name=smoothing,
            line=dict(dash=PLOT_LINESTYLES.get(smoothing, "solid")),
        ))
    fig.update_layout(
        title=f"Precision vs K — {basis}",
        xaxis_title="K (coefficients per arm)",
        yaxis_title="Precision (mean ± std)",
        height=400,
        template="plotly_white",
    )
    fig.show()
```

## Sensitivity (Recall) vs K -- one panel per basis, lines by smoothing method

```python
for basis in bases:
    sub = summary[summary["basis"] == basis].sort_values("n_coeffs_per_arm")
    fig = go.Figure()
    for smoothing, grp in sub.groupby("smoothing"):
        fig.add_trace(go.Scatter(
            x=grp["n_coeffs_per_arm"], y=grp["sensitivity_mean"],
            error_y=dict(type="data", array=grp["sensitivity_std"].values, visible=True),
            mode="lines+markers",
            name=smoothing,
            line=dict(dash=PLOT_LINESTYLES.get(smoothing, "solid")),
        ))
    fig.update_layout(
        title=f"Sensitivity (Recall) vs K — {basis}",
        xaxis_title="K (coefficients per arm)",
        yaxis_title="Sensitivity (mean ± std)",
        height=400,
        template="plotly_white",
    )
    fig.show()
```

## Smoothing vs no-smoothing comparison

For each basis and K, compute the PR-AUC delta (smoothed - none).

```python
baseline = summary[summary["smoothing"] == "none"][["basis", "n_coeffs_per_arm", "pr_auc_mean"]].rename(
    columns={"pr_auc_mean": "pr_auc_baseline"}
)
comparison = summary.merge(baseline, on=["basis", "n_coeffs_per_arm"], how="left")
comparison["pr_auc_delta"] = comparison["pr_auc_mean"] - comparison["pr_auc_baseline"]

smoothed_only = comparison[comparison["smoothing"] != "none"].sort_values("pr_auc_delta", ascending=False)
print("Smoothing configs that improved over no-smoothing baseline:")
display(smoothed_only[smoothed_only["pr_auc_delta"] > 0][
    ["basis", "smoothing", "n_coeffs_per_arm", "pr_auc_mean", "pr_auc_baseline", "pr_auc_delta"]
].head(20))

print("\nSmoothing configs that hurt:")
display(smoothed_only[smoothed_only["pr_auc_delta"] < 0][
    ["basis", "smoothing", "n_coeffs_per_arm", "pr_auc_mean", "pr_auc_baseline", "pr_auc_delta"]
].tail(10))
```
